```python
#Exploratory data analysis
df1["Department"].value_counts()
```

```python
mean_dep_sal = df1.groupby("Department")["MonthlyIncome"].mean().sort_values(ascending=True)
mean_dep_sal

#plot
plt.figure(figsize=(7,7))
mean_dep_sal.plot(kind="bar", color="skyblue")
plt.title("mean monthlyincome by department")
plt.xlabel("Deparment")
plt.ylabel("Monthly income")

plt.show()
```

```python
mean_gend_sal = df1.groupby("Gender")["MonthlyIncome"].mean().sort_values(ascending=True)

#plot
plt.figure(figsize=(7,7))
mean_gend_sal.plot(kind="bar", color="skyblue")
plt.title("mean monthlyincome by gender")
plt.xlabel("gender")
plt.ylabel("Monthly income")

plt.show()
```
