```python
# gender distribution per department
gend_dist_dept = pd.read_sql("SELECT Department,Gender, COUNT(*) numofemployees "
            "FROM employeedata " 
            "GROUP BY Department, Gender "
            "ORDER BY numofemployees;", con=engine)

#plot
sb.set_style("darkgrid")
plt.figure(figsize=(12,7))
sb.catplot(
    data=gend_dist_dept, kind="bar",
    x="Department", y="numofemployees", hue="Gender",
    errorbar="sd", palette="dark", alpha=.6, height=6)
plt.title("gender distribuion per department")
plt.xlabel("Deparment")
plt.ylabel("num of employees")

plt.show()
```

```python
# gender distribution per education field
gend_dist_Edfld=pd.read_sql("SELECT EducationField,Gender, COUNT(*) numofemployees "
            "FROM employeedata " 
            "GROUP BY EducationField, Gender "
            "ORDER BY numofemployees;", con=engine)

#plot
#sb.set_style("darkgrid")
plt.figure(figsize=(12,7))
sb.catplot(
    data=gend_dist_Edfld, kind="bar",
    x="EducationField", y="numofemployees", hue="Gender",
    errorbar="sd", palette="dark", alpha=.6, height=7)
plt.title("gender distribuion by Education Field")
plt.xlabel("Education Field")
plt.ylabel("num of employees")

plt.show()
```

```python
# gender distrbution by Job role
gend_dist_JbRl = pd.read_sql("SELECT JobRole,Gender, COUNT(*) numofemployees "
            "FROM employeedata "
            "GROUP BY JobRole, Gender "
            "ORDER BY numofemployees;", con=engine)

#plot
#sb.set_style("darkgrid")
plt.figure(figsize=(15,7))
sb.barplot(
    data=gend_dist_JbRl,
    x="JobRole", y="numofemployees", hue="Gender",
    errorbar="sd", palette="dark", alpha=.6, width=0.5)
plt.title("gender distribuion by Job Role")
plt.xlabel("Job Role")
plt.ylabel("num of employees")

plt.tight_layout()
plt.show()
```

```python
# Gender disrbution by Job Level
gend_dist_JbLvl=pd.read_sql("SELECT JobLevel,Gender, COUNT(*) numofemployees "
            "FROM employeedata "
            "GROUP BY JobLevel, Gender "
            "ORDER BY JobLevel, Gender, numofemployees;", con=engine)

#plot
#sb.set_style("darkgrid")
plt.figure(figsize=(12,7))
sb.barplot(
    data=gend_dist_JbLvl,
    x="JobLevel", y="numofemployees", hue="Gender",
    errorbar="sd", palette="dark", alpha=.6, width=0.5)
plt.title("gender distribuion by Job Level")
plt.xlabel("Job Level")
plt.ylabel("num of employees")

plt.tight_layout()
plt.show()
```

# **ATTRITION ANALYSIS**
### - Factors influencing employee attrition
### - Compare attrition rates accross different departments, Joblevels etc ...


```python
print(df1["Attrition"].value_counts())
```

```python
df1.loc[df1["Attrition"] == "Yes"].head(10)
```

```python
# Attriion by deparment
attrby_dep = pd.read_sql("SELECT a.Department, COUNT(*) as numofemployees "
                         "FROM (SELECT Department FROM employeedata WHERE Attrition = 'Yes') a "
                         "GROUP BY Department ORDER BY numofemployees;", con=engine)

#plot
sb.set_style("darkgrid")
plt.figure(figsize=(7,7))
sb.barplot(attrby_dep, x="Department", y="numofemployees")
plt.title("Attrition by deparment")
plt.xlabel("Deparment")
plt.ylabel("number of employees")

plt.show()
```

```python
# Attrition by job level
attrby_joblvl = pd.read_sql("SELECT a.JobLevel, COUNT(*) numofemployees "
                            "FROM (SELECT JobLevel FROM employeedata WHERE Attrition = 'Yes') a GROUP BY JobLevel "
                            "ORDER BY JobLevel ASC;", con=engine)

#plot
#sb.set_style("darkgrid")
plt.figure(figsize=(7,7))
sb.barplot(attrby_joblvl, x="JobLevel", y="numofemployees")
plt.title("Attrition by Job Level")
plt.xlabel("Deparment")
plt.ylabel("number of employees")

plt.show()

```

```python
# Attrition against job role
attrby_jobrl = pd.read_sql("SELECT a.JobRole, COUNT(*) numofemployees FROM (SELECT JobRole "
                           "FROM employeedata WHERE Attrition = 'Yes') a "
                           "GROUP BY JobRole ORDER BY numofemployees;", con=engine)

#plot
#sb.set_style("darkgrid")
plt.figure(figsize=(15, 7))
sb.barplot(attrby_jobrl, x="JobRole", y="numofemployees", width=0.5)
plt.title("Attrition by JobRole")
plt.xlabel("Job Role")
plt.ylabel("number of employees")

plt.tight_layout()
plt.show()
```

```python
# Attrition by Age
attrby_age = pd.read_sql("SELECT a.Age, COUNT(*) numofemployees "
                         "FROM (SELECT Age FROM employeedata WHERE Attrition = 'Yes') a "
                         "GROUP BY Age ORDER BY Age;", con=engine)

#plot
#sb.set_style("darkgrid")
plt.figure(figsize=(15, 7))
sb.barplot(attrby_age, x="Age", y="numofemployees", width=0.5)
plt.title("Attrition by age")
plt.xlabel("Age")
plt.ylabel("number of employees")

plt.tight_layout()
plt.show()
```

```python
# Attrition by gender
attrby_gender = pd.read_sql("SELECT a.Gender, COUNT(*) numofemployees "
            "FROM (SELECT Gender FROM employeedata WHERE Attrition = 'Yes') a "
            "GROUP BY Gender "
            "ORDER BY numofemployees;", con=engine)

#plot
#sb.set_style("darkgrid")
plt.figure(figsize=(7, 7))
sb.barplot(attrby_gender, x="Gender", y="numofemployees", width=0.5)
plt.title("Attrition by Gender")
plt.xlabel("Gender")
plt.ylabel("number of employees")

plt.tight_layout()
plt.show()
```

```python
# Attrition by Education field
attrby_Edfld = pd.read_sql("SELECT a.EducationField, COUNT(*) numofemployees "
                           "FROM (SELECT EducationField FROM employeedata WHERE Attrition = 'Yes') a " 
                           "GROUP BY EducationField ORDER BY numofemployees ASC;", con=engine)

#plot
#sb.set_style("darkgrid")
plt.figure(figsize=(10, 7))
sb.barplot(attrby_Edfld, x="EducationField", y="numofemployees", width=0.5)
plt.title("Attrition by Education Field")
plt.xlabel("Education Field")
plt.ylabel("number of employees")

plt.tight_layout()
plt.show()

```

```python
#Attrition by education Level
attrby_Edlvl = pd.read_sql("SELECT a.Education, COUNT(*) numofemployees "
            "FROM (SELECT Education FROM employeedata WHERE Attrition = 'Yes') a "
            "GROUP BY Education ORDER BY Education;", con=engine)

#plot
#sb.set_style("darkgrid")
plt.figure(figsize=(10, 7))
sb.barplot(attrby_Edlvl, x="Education", y="numofemployees", width=0.5)
plt.title("Attrition by Education")
plt.xlabel("Education Level")
plt.ylabel("number of employees")

plt.tight_layout()
plt.show()
```

```python
attrby_ypromotion =pd.read_sql("SELECT a.YearsSinceLastPromotion, COUNT(*) numofemployees "
              "FROM (SELECT YearsSinceLastPromotion FROM employeedata WHERE Attrition = 'Yes') a "
              "GROUP BY YearsSinceLastPromotion "
              "ORDER BY YearsSinceLastPromotion;", con=engine)

#plot
#sb.set_style("darkgrid")
plt.figure(figsize=(12, 7))
sb.barplot(attrby_ypromotion, x="YearsSinceLastPromotion", y="numofemployees", width=0.5)
plt.title("Attrition by Years Since Last Promotion")
plt.xlabel("Years Since Last Promotion")
plt.ylabel("number of employees")

plt.tight_layout()
plt.show()
```

```python
# Attrition by Years at the company
attrby_yatcomp=pd.read_sql("SELECT a.YearsAtCompany, COUNT(*) numofemployees "
            "FROM (SELECT YearsAtCompany FROM employeedata WHERE Attrition = 'Yes') a "
            "GROUP BY YearsAtCompany "
            "ORDER BY YearsAtCompany;", con=engine)

#plot
#sb.set_style("darkgrid")
plt.figure(figsize=(15, 7))
sb.barplot(attrby_yatcomp, x="YearsAtCompany", y="numofemployees", width=0.5)
plt.title("Attrition by Years at company")
plt.xlabel("Years at company")
plt.ylabel("number of employees")

plt.tight_layout()
plt.show()
```

```python
# Attrition by total working years
attrby_Ttwkngy = pd.read_sql("SELECT a.TotalWorkingYears, COUNT(*) numofemployees "
            "FROM (SELECT TotalWorkingYears FROM employeedata WHERE Attrition = 'Yes') a "
            "GROUP BY TotalWorkingYears "
            "ORDER BY TotalWorkingYears;", con=engine)

#plot
#sb.set_style("darkgrid")
plt.figure(figsize=(15, 7))
sb.barplot(attrby_Ttwkngy, x="TotalWorkingYears", y="numofemployees", width=0.5)
plt.title("Attrition by Total working years")
plt.xlabel("Total working years")
plt.ylabel("number of employees")

plt.tight_layout()
plt.show()
```

```python
# Attrition by Years with current manager
attrby_Ywthcurrmng=pd.read_sql("SELECT a.YearsWithCurrManager, COUNT(*) numofemployees "
            "FROM (SELECT YearsWithCurrManager FROM employeedata WHERE Attrition = 'Yes') a "
            "GROUP BY YearsWithCurrManager "
            "ORDER BY YearsWithCurrManager;", con=engine)

#plot
#sb.set_style("darkgrid")
plt.figure(figsize=(15, 7))
sb.barplot(attrby_Ywthcurrmng, x="YearsWithCurrManager", y="numofemployees", width=0.5)
plt.title("Attrition by Years with current Manager")
plt.xlabel("Years With current manager")
plt.ylabel("number of employees")

plt.tight_layout()
plt.show()
```

```python
Avg_dep_sal = pd.read_sql("SELECT Department, AVG(MonthlyIncome) AvgMonthlyincome "
            "FROM employeedata "
            "GROUP BY Department "
            "ORDER BY AvgMonthlyincome;", con=engine)

#plot
#sb.set_style("darkgrid")
plt.figure(figsize=(7, 7))
sb.barplot(Avg_dep_sal, x="Department", y="AvgMonthlyincome", width=0.5)
plt.title("Average Monthly Income by Department")
plt.xlabel("Department")
plt.ylabel("Monthly income")

plt.tight_layout()
plt.show()
```

```python
df1["EducationField"].value_counts()
```

```python
Avg_EF_sal = pd.read_sql("SELECT EducationField, AVG(MonthlyIncome) avg_monthly_income "
            "FROM employeedata "
            "GROUP BY EducationField "
            "ORDER BY avg_monthly_income;", con=engine)


#plot
#sb.set_style("darkgrid")
plt.figure(figsize=(10, 7))
sb.barplot(Avg_EF_sal, x="EducationField", y="avg_monthly_income", width=0.5)
plt.title("Average Monthly Income by Education Field")
plt.xlabel("Education Field")
plt.ylabel("Monthly income")

plt.tight_layout()
plt.show()
```

```python
# average monthly income by job role
Avg_JR_sal=pd.read_sql("SELECT JobRole, AVG(MonthlyIncome) avg_monthly_income "
            "FROM employeedata "
            "GROUP BY JobRole "
            "ORDER BY avg_monthly_income;", con=engine)


#plot
#sb.set_style("darkgrid")
plt.figure(figsize=(15, 6))
sb.barplot(Avg_JR_sal, x="JobRole", y="avg_monthly_income", width=0.5)
plt.title("Average Monthly Income by Job Role")
plt.xlabel("Job Role")
plt.ylabel("Monthly income")

plt.tight_layout()
plt.show()
```

```python
Avg_JL_sal = pd.read_sql("SELECT JobLevel, AVG(MonthlyIncome) avg_monthly_income "
            "FROM employeedata "
            "GROUP BY JobLevel "
            "ORDER BY JobLevel;", con=engine)

#plot
#sb.set_style("darkgrid")
plt.figure(figsize=(10, 6))
sb.barplot(Avg_JL_sal, x="JobLevel", y="avg_monthly_income", width=0.5)
plt.title("Average Monthly Income by Job Level")
plt.xlabel("Job Level")
plt.ylabel("Monthly income")

plt.tight_layout()
plt.show()
```
